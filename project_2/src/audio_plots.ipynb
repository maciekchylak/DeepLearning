## Plots to show how audio data looks like

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile as wav
from python_speech_features import mfcc
```

### Bed

```python
sample_rate, sample = wav.read(f"project_2/train/audio/bed/0a7c2a8d_nohash_0.wav")

length_of_audio = len(sample) / sample_rate
time = np.linspace(0, length_of_audio, len(sample))

plt.plot(time, sample)

plt.ylabel("Amplitude")
plt.xlabel("Time")

plt.show()
```

```python
_ = plt.specgram(sample, Fs=sample_rate)

plt.ylabel("Frequency")
plt.xlabel("Time")

plt.show()
```

```python
mfccc = mfcc(sample, sample_rate)
time = np.linspace(0, length_of_audio, len(sample))
```

```python
plt.figure(figsize=(10,16))
plt.imshow(mfccc.T, origin='lower')

plt.xlabel("Time")
plt.ylabel("MFCC")
```
